# ZenQuotes

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import time
```

# Keywords

```python
keywords_url = "https://zenquotes.io/keywords"
keywords_request = requests.get(keywords_url)
keywords_html = BeautifulSoup(keywords_request.text)
print(keywords_html)
```

## Getting all the keywords available

```python
target = keywords_html.find("h2",text="Browse Keywords")
for sib in target.find_next_siblings():
    if sib.name == "h2":
        break
    else:
        k = sib.text
keys = k.split()
print(keys)
```

# Getting quotes for every keyword

```python
def get_quotes_for_one_key(keyword):
    authors = []
    quotes = []
    keys = []
    length = []
    link = "https://zenquotes.io/api/quotes/" + keyword
    response = requests.get(link)
    json_data = json.loads(response.text)
    for i in range(len(json_data)):
        authors.append(json_data[i]["a"])
        quotes.append(json_data[i]["q"])
        length.append(json_data[i]["c"])
        keys.append(keyword)
        time.sleep(8)
    df = pd.DataFrame()
    df["author_name"] = authors
    df["quotes"] = quotes
    df["char_length"] = length
    df["keyword"] = keys
    return df
```

```python
j = 0
for i in keys:
    i = get_quotes_for_one_key(i)
    i.to_csv(f"df_{j}.csv")
    print(f"df_{j}.csv saved")
    j += 1
```
